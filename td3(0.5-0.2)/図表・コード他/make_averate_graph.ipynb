```python
import matplotlib.pyplot as plt

# 
with open('results/averages.txt', 'r') as f:  # 'data.txt'
    lines = f.readlines()

# 
values = [float(line.strip()) for line in lines]

# 
fig_reward = plt.figure(dpi = 300)  # 
plt.plot(range(1, len(values)+1), values)
plt.xlabel('episode')
plt.ylabel('rewards')
plt.title('Average Rewards per Episode')

# 
plt.tight_layout()

# png
fig_reward.savefig('results/average_graph.jpg', format="jpg", bbox_inches='tight')  # 'graph.png'
plt.show()
plt.close()
```
