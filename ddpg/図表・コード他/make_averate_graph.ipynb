```python
import matplotlib.pyplot as plt

# ファイルを読み込む
with open('results/averages.txt', 'r') as f:  # 'data.txt'を適切なファイル名に書き換えてください
    lines = f.readlines()

# 数値のリストを作成
values = [float(line.strip()) for line in lines]

# グラフの描画
fig_reward = plt.figure(dpi = 300)  # グラフのサイズを指定（適宜変更可能）
plt.plot(range(1, len(values)+1), values)
plt.xlabel('episode')
plt.ylabel('rewards')
plt.title('Average Rewards per Episode')

# レイアウトの調整
plt.tight_layout()

# png形式で保存
fig_reward.savefig('results/average_graph.jpg', format="jpg", bbox_inches='tight')  # 'graph.png'を希望のファイル名に書き換えてください
plt.show()
plt.close()
```
