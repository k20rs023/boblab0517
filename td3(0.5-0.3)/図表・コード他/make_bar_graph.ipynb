```python
import matplotlib.pyplot as plt
import pandas as pd

# データの読み込み
df = pd.read_csv('results/all_output.txt', header=None, names=['values'])

# ビニング
bins = range(df['values'].min(), df['values'].max() + 1000, 1000)
df['binned'] = pd.cut(df['values'], bins)

# 分布の計算と棒グラフの描画
df['binned'].value_counts().sort_index().plot(kind='bar')

# グラフを保存する前にレイアウトを調整
plt.tight_layout()

# グラフの保存
plt.savefig('results/bar_graph.jpg', format='jpg', dpi=300)

# グラフの表示
plt.show()
```
